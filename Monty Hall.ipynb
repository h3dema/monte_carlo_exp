```python
# Monty Hall Problem
```

Suponha que você esteja em um game show e tenha a opção de escolher uma das três portas:
Atrás de uma porta está um carro; atrás das outras portas, cabras. 
Você escolhe uma porta, digamos a porta 1, 
O apresentado, que sabe o que está atrás das portas, abre outra porta, digamos a porta 3, que tem uma cabra.
Ainda com a porta 1 fechada, o anfitrião então pergunta: você quer manter sua escolha ou escolher outra porta?
Você fica com a porta 1 ou escolhe a porta 2 para ganhar o carro?


Este problema é interessante porque normalmente não estamos acostumados a pensar em probabilidades e como isto afeta o resultado.

__Com base na probabilidade, descobrimos que é uma vantagem para nós trocar as portas.__

Inicialmente, a probabilidade (P) de pegar o carro é a mesma (P = 1/3) para todas as portas.

Agora suponha que o competidor escolha a porta 1. Em seguida, o anfitrião abre a terceira porta, que tem uma cabra. Em seguida, o apresentador pergunta ao competidor se ele deseja trocar as portas?

Após o anfitrião abrir a porta 3, a probabilidade das duas últimas portas terem um carro aumenta para 2/3. Agora que sabemos que a terceira porta tem uma cabra, a probabilidade da segunda porta ter um carro aumenta para 2/3. Portanto, é mais vantajoso trocar as portas.

__Parece estranho, não é?__

```python
import time
import random
import matplotlib.pyplot as plt

from IPython.display import clear_output
```

```python
# inicialização

portas = ["cabra", "cabra", "carro"]
```

```python
# probabilidades

prob_mudar = []
prob_ficar = []

ganha_mudar = 0
ganha_ficar = 0

N = 100

for j in range(N):
    clear_output(wait=True)

    # vamos alterar aleatoriamente a posicao do que está atrás de cada porta
    random.shuffle(portas)
    # e em seguida vamos fazer nossa escolha aleatoriamente
    porta = random.randint(0, 2)  # retorna 0, 1 ou 2
   
    
    if portas[porta] == "carro":
        ganha_ficar += 1
    else:
        ganha_mudar += 1
        
    prob_ficar.append(ganha_ficar / (j + 1))
    prob_mudar.append(ganha_mudar / (j + 1))
    
    
    fig, _ = plt.subplots(1, 1, figsize=(16, 9))
    
    
    plt.axhline(y=2.0/3.0, color="green", linestyle="-")  # mudar
    plt.text(len(prob_mudar)-1, 2.0/3.0 + 0.02, "Mudar")
    plt.axhline(y=1.0/3.0 + 0.01, color="red", linestyle="-")  # ficar
    plt.text(len(prob_mudar)-1, 1.0/3.0 + 0.02, "Ficar")
        
    plt.xlabel("Iteracao")
    plt.plot(prob_ficar, color="orangered")
    plt.plot(prob_mudar, color="limegreen")
    
    plt.show()
    time.sleep(0.1)
    plt.close(fig)
    
print(f"Probabilidade de ganhar se ficar com a primeira escolha: {prob_ficar[-1]:.3f}")
print(f"Probabilidade de ganhar se trocar de porta             : {prob_mudar[-1]:.3f}")
```
