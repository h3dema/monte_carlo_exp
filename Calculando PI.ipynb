```python
import time
import matplotlib.pyplot as plt
import random
import numpy as np

from IPython.display import clear_output
```

```python
def dist(p):
    return np.linalg.norm(p)
```

$\frac{area do circulo}{area do quadrado} = \frac{\pi \times r^2}{2r \times 2r} = \frac{\pi}{4}$

$\pi = 4 * \frac{numero total de pontos dentro do circulo}{total de pontos no quadrado}$

```python
points = []
N = 5000  # number of points to add each cycle
for i in range(5):
    fig, ax = plt.subplots(figsize=(9,9))
    
    circle = plt.Circle((0, 0), 1, color='yellow', alpha=0.1)
    ax.add_patch(circle)
    
    points.extend([(random.random(), random.random()) for _ in range(N)])
    num_inside = 0
    for p in points:
        inside = dist(p) <= 1
        plt.plot(p[0], p[1], '.', color='red' if inside else "blue")
        if inside:
            num_inside += 1
            
    _pi = 4 * num_inside / len(points)
    plt.title(f"N#{len(points)} --> PI = {_pi}")
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.show()
    time.sleep(0.5)
    clear_output(wait=True)
    plt.close(fig)
```
